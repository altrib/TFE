```python
import numpy as np

import netCtest4 as nc
from datetime import datetime, timedelta, time


import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import pandas as pd
import os

from scipy import stats
import scipy as sp

import seaborn as sns
```

```python
%matplotlib widget
test = pd.read_csv("./notebooks/test.csv",sep=',')

test = test[abs(test['tail'])<1]


# Create a figure with subplots
fig = plt.figure(figsize=(15, 10))
plt.title("Dependence between parameter estimations\non Schiphol RACMO, BM bootstrap")
# 3D scatter plot
ax_3d = fig.add_subplot(221, projection='3d')
ax_3d.scatter(test['tail'], test['scale'], test['loc'], c='b', marker='o')
ax_3d.set_xlabel('tail')
ax_3d.set_ylabel('scale')
ax_3d.set_zlabel('loc')

# Joint density plot for x and y
ax_xy = fig.add_subplot(222)
sns.kdeplot(x=test['tail'], y=test['scale'], fill=True, ax=ax_xy, cmap="Blues")
ax_xy.set_xlabel('tail')
ax_xy.set_ylabel('scale')

# Joint density plot for x and z
ax_xz = fig.add_subplot(223)
sns.kdeplot(x=test['tail'], y=test['loc'], fill=True, ax=ax_xz, cmap="Greens")
ax_xz.set_xlabel('tail')
ax_xz.set_ylabel('loc')

# Joint density plot for y and z
ax_yz = fig.add_subplot(224)
sns.kdeplot(x=test['scale'], y=test['loc'], fill=True, ax=ax_yz, cmap="Reds")
ax_yz.set_xlabel('scale')
ax_yz.set_ylabel('loc')

# Adjust layout
plt.tight_layout()
plt.show()
```
